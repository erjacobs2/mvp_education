```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#import csv
df = pd.read_csv('Cleaned_Data.csv')
df.head()
```

```python
# Assign the data to X and y
# Note: Sklearn requires a two-dimensional array of values
# so we use reshape to create this

# print("Shape: ", X.shape, math_4.shape)
X = df.TOTAL_REVENUE.values.reshape(-1, 1)
y = df.AVG_MATH_4_SCORE.values.reshape(-1, 1)

south = df.loc[df['Region'] == 'South']
west = df.loc[df['Region'] == 'West']
northeast = df.loc[df['Region'] == 'Northeast']
midwest = df.loc[df['Region'] == 'Midwest']

X_south = south.TOTAL_REVENUE.values.reshape(-1, 1)
math4south = south.AVG_MATH_4_SCORE.values.reshape(-1, 1)

X_west = west.TOTAL_REVENUE.values.reshape(-1, 1)
math4west = west.AVG_MATH_4_SCORE.values.reshape(-1, 1)

X_northeast = northeast.TOTAL_REVENUE.values.reshape(-1, 1)
math4ne = northeast.AVG_MATH_4_SCORE.values.reshape(-1, 1)

X_midwest = midwest.TOTAL_REVENUE.values.reshape(-1, 1)
math4mw = midwest.AVG_MATH_4_SCORE.values.reshape(-1, 1)
```

```python
plt.scatter(X_south, math4south)
# blue
plt.scatter(X_west, math4west)
# orange
plt.scatter(X_northeast, math4ne)
# green
plt.scatter(X_midwest, math4mw) 
# red
```

```python
# Create the model and fit the model to the data

from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
print('Weight coefficients: ', model.coef_)
print('y-axis intercept: ', model.intercept_)
```

```python
x_min = np.array([[X.min()]])
x_max = np.array([[X.max()]])
print(f"Min X Value: {x_min}")
print(f"Max X Value: {x_max}")
```

```python
y_min = model.predict(x_min)
y_max = model.predict(x_max)
```

```python
s = plt.scatter(X_south, math4south, c='blue')
w = plt.scatter(X_west, math4west, c='orange')
ne = plt.scatter(X_northeast, math4ne, c='green')
mw = plt.scatter(X_midwest, math4mw, c='red')

plt.plot([x_min[0], x_max[0]], [y_min[0], y_max[0]], c='black')

plt.xlabel('Total Revenue (Tens of Millions)')
plt.ylabel('Avg. 4th Grade Math Scores')
plt.title('Revenue vs. 4th Grade Math Scores')
plt.show()
```

```python
plt.savefig("Images/Math4.png")
```
