```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
cleanedData_to_load = "Data/Cleaned_Data.csv"
spend_df = pd.read_csv(cleanedData_to_load)
spend_df.columns
```

```python
spend_region = spend_df.groupby(['Region'])
avg_rev_ps_region = round(spend_region["TOTAL_REVENUE_PS"].mean(),2)
avg_spend_ps_region = round(spend_region["TOTAL_EXPENDITURE_PS"].mean(),2)
avg_spend_instruction_region = round(spend_region["INSTRUCTION_EXPENDITURE_PS"].mean(),2)
avg_spend_support_ps_region = round(spend_region["SUPPORT_SERVICES_EXP_PS"].mean(),2)
avg_spend_other_ps_region = round(spend_region["OTHER_EXPENDITURE_PS"].mean(),2)
avg_spend_capX_ps_region = round(spend_region["CAPITAL_OUTLAY_EXP_PS"].mean(),2)
```

```python
#create a dataframe to store my calculated datahttp://localhost:8888/notebooks/MVP_SpendByRegion.ipynb#
spend_region_df = pd.DataFrame({"Average Revenue Per Student":avg_rev_ps_region,
                 "Average Spend Per Student": avg_spend_ps_region,
                 "Average Instructional Spend Per Student": avg_spend_instruction_region,
                 "Average Support Servces Spend Per Student": avg_spend_support_ps_region, 
                 "Average Other Spend Per Student": avg_spend_other_ps_region,
                 "Average Capital Expenditure Per Student": avg_spend_capX_ps_region
})
spend_region_df
```

```python
spend_state = spend_df.groupby(['STATE'])
avg_rev_ps_state = round(spend_state["TOTAL_REVENUE_PS"].mean(),2)
avg_spend_ps_state = round(spend_state["TOTAL_EXPENDITURE_PS"].mean(),2)
avg_spend_instruction_state = round(spend_state["INSTRUCTION_EXPENDITURE_PS"].mean(),2)
avg_spend_support_ps_state = round(spend_state["SUPPORT_SERVICES_EXP_PS"].mean(),2)
avg_spend_other_ps_state = round(spend_state["OTHER_EXPENDITURE_PS"].mean(),2)
avg_spend_capX_ps_state = round(spend_state["CAPITAL_OUTLAY_EXP_PS"].mean(),2)
```

```python
spend_state_df = pd.DataFrame({"Average Revenue Per Student":avg_rev_ps_state,
                 "Average Spend Per Student": avg_spend_ps_state,
                 "Average Instructional Spend Per Student": avg_spend_instruction_state,
                 "Average Support Servces Spend Per Student": avg_spend_support_ps_state, 
                 "Average Other Spend Per Student": avg_spend_other_ps_state,
                 "Average Capital Expenditure Per Student": avg_spend_capX_ps_state
})

spend_state_df
```

```python
x_axis = np.arange(len(spend_state_df))
tick_locations = [value+0.2 for value in x_axis]
```

```python
plt.figure(figsize=(20,3))
plt.bar(x_axis, spend_state_df["Average Spend Per Student"], color='r', alpha=0.5, align="edge")
```
